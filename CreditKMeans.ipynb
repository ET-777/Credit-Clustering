```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

<a name="2"></a>
## 2 - Preprocessing Data

The dataset contains information about customers and their credit cards, as well as their interaction with the bank.

The dataset can be found here: [Credit Card Data](https://www.kaggle.com/datasets/aryashah2k/credit-card-customer-data)
<br/><br/>
<a name="2.1"></a>
### 2.1 Loading and Visualizing the Data

```python
#Laod data
data = pd.read_csv("./Data/Credit Card Customer Data.csv")
```

```python
data.head()
```

```python
print(f"Shape: {data.shape}")
```

<a name="2.2"></a>
### 2.2 Preparing Data

Let's get rid of the unecessary columns to fit our data, and create y variable for graphing.

```python
#Get features
x = data.drop(['Sl_No', 'Customer Key'], axis=1).to_numpy()
```

```python
#Create y variables
y_1 = data['Total_Credit_Cards'].to_numpy()
y_2 = data['Total_visits_bank'].to_numpy()
y_3 = data['Total_visits_online'].to_numpy()
y_4 = data['Total_calls_made'].to_numpy()
```

<a name="3"></a>
## 3 - Clustering

<a name="3.1"></a>
### 3.1 K Means

We'll fit our data to 2 clusters using KMeans from scikit-learn.

```python
#Fit data
kmeans = KMeans(n_clusters=2, random_state=0).fit(x)
```

```python
print(f"Predictions: {kmeans.labels_}\n")
print(f"Cluster centers [x,y]: {kmeans.cluster_centers_}")
```

Above, we can see to which cluster our items group to, and the centers for each cluster.

<a name="3.2"></a>
### 3.2 Visualizing Clusters

Now let's graph our findings to see to which group the customers belong to with each category.

```python
centroids = kmeans.cluster_centers_
```

```python
#Plot graphs
plt.subplots(figsize=(20,15))
plt.subplot(2, 2, 1)
plt.title('Credit Limit vs Credit Cards')
plt.scatter(x[:, 0], y_1, c=kmeans.labels_)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=160, linewidths=3, color="r")
plt.xlabel('Credit Limit')
plt.ylabel('Credit Cards')

plt.subplot(2, 2, 2)
plt.title('Credit Limit vs Bank Visits')
plt.scatter(x[:, 0], y_2, c=kmeans.labels_)
plt.scatter(centroids[:, 0], centroids[:, 2], marker="x", s=160, linewidths=3, color="r")
plt.xlabel('Credit Limit')
plt.ylabel('Bank Visits')

plt.subplot(2, 2, 3)
plt.title('Credit Limit vs Online Visits')
plt.scatter(x[:, 0], y_3, c=kmeans.labels_)
plt.scatter(centroids[:, 0], centroids[:, 3], marker="x", s=160, linewidths=3, color="r")
plt.xlabel('Credit Limit')
plt.ylabel('Online Visits')

plt.subplot(2, 2, 4)
plt.title('Credit Limit vs Calls')
plt.scatter(x[:, 0], y_4, c=kmeans.labels_)
plt.scatter(centroids[:, 0], centroids[:, 4], marker="x", s=160, linewidths=3, color="r")
plt.xlabel('Credit Limit')
plt.ylabel('Calls')
```

<a name="4"></a>
## 4 - Results

We can see clear cluster groups in the graphs. The centers are marked by the red crosses. From these we can observe:

- Customers with a credit higher than 75000 are high credit customers.
- Customer with high credit tend to have more credit cards.
- Customers with a lower credit have more interactions with the bank: calls and visits.
- Customers with high credit have more online visits.

These findings seem to indicate high credit customers have an easier experience with the bank.
